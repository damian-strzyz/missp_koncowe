# **Zadanie 1**


*   Wypisanie zajęć od środy do piątku
*   Wypisanie zajęć w dniach innych niż środa



```python
## Dane:
zajecia = ["wf","informatyka","relax","mispy","piateczek"]
dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

## Rozwiązanie:
## Część pierwsza:
for dzien , lekcja in enumerate(zajecia):
    if dzien > 1:
        print("Zajęcia w {} to {}".format(dnityg[dzien] , lekcja))

print("-------------")

## Część druga:
for dzien , lekcja in enumerate(zajecia):
    if not dzien == 2:
        print("Zajęcia w {} to {}".format(dnityg[dzien] , lekcja))
```

# **Zadanie 2**


*   Zdefiniowanie funkcji która poda ilość zajęć w danym dniu
*   Posortowanie listy rosnąco po ilości zajęć



```python
## Dane:
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie", "spanie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

## Część pierwsza:
def liczba_zajec(dnityg, zajecia):
    ile_zaj =  [len(x) for x in zajecia]
    ile_zaj = list(zip(dnityg, ile_zaj))
    return ile_zaj
ilosc_zaj = liczba_zajec(dnityg, zajecia) 
print(ilosc_zaj)

## Część druga:
wez_drugi = lambda krotka: krotka[1]
sorted(ilosc_zaj, key = wez_drugi)
```

# **Zadanie 3**


*   Program znajdujący dzień z największą ilością mispów




```python
## Dane:
zajecia = [["wf"],
           ["informatyka", "nauka czytania"],
           ["relax", "jedzenie"],
           ["mispy","mispy w domu"],
           ["piateczek", "praca domowa z mispow", "Ilovemispy", "mispy w czasie snu"]
          ]

dnityg= ["Pon","Wto","Sro","Czwar","Pio"]

## Program:
d=0
ile_d=[0,0,0,0,0]
for value in zajecia:
    n=0
    for value1 in value:
        if value1.find("misp") >= 0:
            n+=1
    ile_d[d]=n
    d+=1
lista = list(zip(dnityg, ile_d))
wez_drugi = lambda krotka: krotka[1]
max(lista, key = wez_drugi)
```
