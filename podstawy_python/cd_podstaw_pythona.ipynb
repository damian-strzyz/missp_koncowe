*   Stworzenie typu NumerTelefonu
*   Podczas tworzenia ma sprawdzić czy numer jest poprawny (Polski), jeżeli nie zwraca ERROR
*   Wypisuje numer słownie jako podstawowa funkcja print typu NumerTelefonu
*   Podczas wywołania właściwości podaje województwo, lub to czy numer jest komórkowy

 
 

```python
## Część pierwsza:
class NumerTelefonu():
    ## Część druga:
    def __init__(self, numer): #definicja
        if len(numer) != 9:
            return print("ERROR")
        else:
            self.numertel= numer
            return print("prawidlowy numer")
    ## Część czwarta:
    @property
    def wlasciwosci(self):
        if "42" in self.numertel:
            return print("wojewodztwo Łódzkie")
        elif "22" in self.numertel:
            return print("wojewodztwo Mazowieckie")
        elif "71" in self.numertel:
            return print("wojewodztwo Śląskie")
        elif "12" in self.numertel:
            return print("wojewodztwo Małopolskie")
        else:
            return print("Telefon komórkowy")

    ## Część trzecia:   
    def __repr__(self):
        odczyt = ""
        pisemnie = ["zero ","jeden ", "dwa ","trzy ","cztery ","pięć ","sześć ","siedem ","osiem ","dziewięć "]
        for a in range(len(self.numertel)):
            odczyt = odczyt + pisemnie[int(self.numertel[a])]
        return odczyt
        
```

```python
numerb = NumerTelefonu("505050505")
```
