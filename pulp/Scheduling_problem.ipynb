# **Zadanie**
Zadanie polega na stworzeniu problemu minimalizacji kosztu tygodniowego zatrudnienia pracowników w sklepie. Grafik jest tworzony przy pomocy funkcji minimalizacji i jest ograniczany przez minimalna liczbe osób które muszą być w pracy każdego dnia, a także możliwości pracowników.

*   Stworzenie słowników dniowych zawierających możliwości do pracy pracowników w poszczególnych dniach, a także ich stawki dniowe
*   Stworzenie celu procesu minimalizacji
*   Dodanie minimalnej liczby pracowników w danym dniu tygodnia
*   Dodanie ograniczeń, aby dany pracownik nie mógłbyć uwzględniony w danym dniu kilkukrotnie
*   Rozwiązanie problemu
*   Wypisanie grafiku tygodniowego, a także łącznego tygodniowego kosztu



```
from pulp import *

Pracownicy = ['Ania', 'Stefan', 'Hektor', 'Olaf', 'Lidia', 'Piotr']

poniedziałek = {
    'Ania': 1,
    'Stefan': 1,
    'Hektor': 1,
    'Olaf': 1,
    'Lidia': 0,
    'Piotr': 1
}
wtorek = {
    'Ania': 1,
    'Stefan': 0,
    'Hektor': 1,
    'Olaf': 1,
    'Lidia': 0,
    'Piotr': 1
}
środa = {
    'Ania': 1,
    'Stefan': 0,
    'Hektor': 1,
    'Olaf': 1,
    'Lidia': 0,
    'Piotr': 1
}
czwartek = {
    'Ania': 1,
    'Stefan': 0,
    'Hektor': 0,
    'Olaf': 1,
    'Lidia': 1,
    'Piotr': 0
}
piątek = {
    'Ania': 1,
    'Stefan': 0,
    'Hektor': 0,
    'Olaf': 1,
    'Lidia': 1,
    'Piotr': 0
}
płaca = {
    'Ania': 100,
    'Stefan': 50,
    'Hektor': 60,
    'Olaf': 40,
    'Lidia': 110,
    'Piotr': 70
}

prob = LpProblem("Sklep_praca",LpMinimize)
pracownicy_varsp = LpVariable.dicts("Praca_Pon",Pracownicy,0)
pracownicy_varsw = LpVariable.dicts("Praca_Wto",Pracownicy,0)
pracownicy_varss = LpVariable.dicts("Praca_Sro",Pracownicy,0)
pracownicy_varsc = LpVariable.dicts("Praca_Czw",Pracownicy,0)
pracownicy_varspi = LpVariable.dicts("Praca_Pia",Pracownicy,0)

prob += lpSum([płaca[i]*(pracownicy_varsp[i]+pracownicy_varsw[i]+pracownicy_varss[i]+pracownicy_varsc[i]+pracownicy_varspi[i]) for i in Pracownicy]),"Koszt_tygodniowy"
prob += lpSum([poniedziałek[i]*pracownicy_varsp[i] for i in Pracownicy]) == 2,"Pracownicy_na_poniedziałek"
prob += lpSum([wtorek[i]*pracownicy_varsw[i] for i in Pracownicy]) == 1,"Pracownicy_na_wtorek"
prob += lpSum([środa[i]*pracownicy_varss[i] for i in Pracownicy]) == 1,"Pracownicy_na_środę"
prob += lpSum([czwartek[i]*pracownicy_varsc[i] for i in Pracownicy]) == 1,"Pracownicy_na_czwartek"
prob += lpSum([piątek[i]*pracownicy_varspi[i] for i in Pracownicy]) == 3,"Pracownicy_na_piątek"

prob += lpSum([(pracownicy_varsp[i]+pracownicy_varsw[i]+pracownicy_varss[i]+pracownicy_varsc[i]+pracownicy_varspi[i]) for i in Pracownicy]) == 8 , "Suma"

prob += (pracownicy_varsp['Ania']+pracownicy_varsw['Ania']+pracownicy_varss['Ania']+pracownicy_varsc['Ania']+pracownicy_varspi['Ania']) <= 3, "A"
prob += (pracownicy_varsp['Hektor']+pracownicy_varsw['Hektor']+pracownicy_varss['Hektor']+pracownicy_varsc['Hektor']+pracownicy_varspi['Hektor']) <= 3, "H"
prob += (pracownicy_varsp['Piotr']+pracownicy_varsw['Piotr']+pracownicy_varss['Piotr']+pracownicy_varsc['Piotr']+pracownicy_varspi['Piotr']) <= 3, "P"
prob += (pracownicy_varsp['Lidia']+pracownicy_varsw['Lidia']+pracownicy_varss['Lidia']+pracownicy_varsc['Lidia']+pracownicy_varspi['Lidia']) <= 3, "L"
prob += (pracownicy_varsp['Olaf']+pracownicy_varsw['Olaf']+pracownicy_varss['Olaf']+pracownicy_varsc['Olaf']+pracownicy_varspi['Olaf']) <= 3, "O"
prob += (pracownicy_varsp['Stefan']+pracownicy_varsw['Stefan']+pracownicy_varss['Stefan']+pracownicy_varsc['Stefan']+pracownicy_varspi['Stefan']) <= 3, "S"

prob += pracownicy_varsp['Ania'] <=1, "A1"
prob += pracownicy_varsw['Ania'] <=1, "A2"
prob += pracownicy_varss['Ania'] <=1, "A3"
prob += pracownicy_varsc['Ania'] <=1, "A4"
prob += pracownicy_varspi['Ania'] <=1, "A5"

prob += pracownicy_varsp['Piotr'] <=1, "P1"
prob += pracownicy_varsw['Piotr'] <=1, "P2"
prob += pracownicy_varss['Piotr'] <=1, "P3"
prob += pracownicy_varsc['Piotr'] <=1, "P4"
prob += pracownicy_varspi['Piotr'] <=1, "P5"

prob += pracownicy_varsp['Hektor'] <=1, "H1"
prob += pracownicy_varsw['Hektor'] <=1, "H2"
prob += pracownicy_varss['Hektor'] <=1, "H3"
prob += pracownicy_varsc['Hektor'] <=1, "H4"
prob += pracownicy_varspi['Hektor'] <=1, "H5"

prob += pracownicy_varsp['Lidia'] <=1, "L1"
prob += pracownicy_varsw['Lidia'] <=1, "L2"
prob += pracownicy_varss['Lidia'] <=1, "L3"
prob += pracownicy_varsc['Lidia'] <=1, "L4"
prob += pracownicy_varspi['Lidia'] <=1, "L5"

prob += pracownicy_varsp['Stefan'] <=1, "S1"
prob += pracownicy_varsw['Stefan'] <=1, "S2"
prob += pracownicy_varss['Stefan'] <=1, "S3"
prob += pracownicy_varsc['Stefan'] <=1, "S4"
prob += pracownicy_varspi['Stefan'] <=1, "S5"

prob += pracownicy_varsp['Olaf'] <=1, "O1"
prob += pracownicy_varsw['Olaf'] <=1, "O2"
prob += pracownicy_varss['Olaf'] <=1, "O3"
prob += pracownicy_varsc['Olaf'] <=1, "O4"
prob += pracownicy_varspi['Olaf'] <=1, "O5"

prob.solve()
print("Status:", LpStatus[prob.status])
f=0
for i in prob.variables():
  print(i.name," ",i.varValue)
  if (f % 6 == 5):
    print("------------------------")
  f += 1 

print("Łączny koszt tygodniowy",value(prob.objective))
```
