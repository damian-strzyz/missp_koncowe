# **Zadanie**

*   Dodanie zmiennej decyzyjnej do powyższego przykładu, który znajdzie się w końcowym składzie
*   Dodanie ograniczenia do zawartości witaminy B
*   Wypisać odpowiedź w formie: Końcowy produkt będzie kosztować {}, będzie miał w składzie {nazwa}: {} %, {nastepna nazwa}: {} %

# **Dane:**

| Nazwa      | zawartość ug/100g |
|------------|-------------------|
| chicken    | 0.3               |
| beef       | 2.6               |
| mutton     | 2.6               |
| rice       | 0                 |
| wheat bran | 0                 |
| gel        | 0                 |
| sardines   | 8.9               |

Kot ma zjeść minimum 0.24 ug na kilogram masy więc około
0.8 ug dziennie


```
"""
The Simplified Whiskas Model Python Formulation for the PuLP Modeller

Authors: Antony Phillips, Dr Stuart Mitchell  2007
"""
from pulp import *

# Creates a list of the Ingredients
Ingredients = ['CHICKEN', 'BEEF', 'MUTTON', 'RICE', 'WHEAT', 'GEL','SARDINES','CORN']

# A dictionary of the costs of each of the Ingredients is created
costs = {'CHICKEN': 0.013, 
         'BEEF': 0.020, 
         'MUTTON': 0.016, 
         'RICE': 0.008, 
         'WHEAT': 0.005, 
         'GEL': 0.001,
          'SARDINES': 0.018,
          'CORN': 0.002}

# A dictionary of the protein percent in each of the Ingredients is created
proteinPercent = {'CHICKEN': 0.100, 
                  'BEEF': 0.200, 
                  'MUTTON': 0.150, 
                  'RICE': 0.000, 
                  'WHEAT': 0.040, 
                  'GEL': 0.000,
                  'SARDINES': 0.250,
                  'CORN': 0.021}

# A dictionary of the fat percent in each of the Ingredients is created
fatPercent = {'CHICKEN': 0.080, 
              'BEEF': 0.100, 
              'MUTTON': 0.110, 
              'RICE': 0.010, 
              'WHEAT': 0.010, 
              'GEL': 0.000,
              'SARDINES': 0.110,
              'CORN': 0.008}

# A dictionary of the fibre percent in each of the Ingredients is created
fibrePercent = {'CHICKEN': 0.001, 
                'BEEF': 0.005, 
                'MUTTON': 0.003, 
                'RICE': 0.100, 
                'WHEAT': 0.150, 
                'GEL': 0.000,
                'SARDINES': 0.000,
                'CORN': 0.081}

# A dictionary of the salt percent in each of the Ingredients is created
saltPercent = {'CHICKEN': 0.002, 
               'BEEF': 0.005, 
               'MUTTON': 0.007, 
               'RICE': 0.002, 
               'WHEAT': 0.008, 
               'GEL': 0.000,
                'SARDINES': 0.005,
               'CORN': 0.000}

# A dictionary of the vit B ug in each of the Ingredients is created
Bvitug = {'CHICKEN': 0.003, 
               'BEEF': 0.010, 
               'MUTTON': 0.010, 
               'RICE': 0.000, 
               'WHEAT': 0.000, 
               'GEL': 0.000,
               'SARDINES': 0.089,
                'CORN': 0.006}

prob = LpProblem("The_Whiskas_Problem",LpMinimize)

ingredient_vars = LpVariable.dicts("Ingr",Ingredients,0)

prob += lpSum([costs[i]*ingredient_vars[i] for i in Ingredients]), "Total_Cost_of_Ingredients_per_can"
prob += lpSum([ingredient_vars[i] for i in Ingredients]) == 100, "PercentagesSum"
prob += lpSum([proteinPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 8.0, "ProteinRequirement"
prob += lpSum([fatPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 6.0, "FatRequirement"
prob += lpSum([fibrePercent[i] * ingredient_vars[i] for i in Ingredients]) <= 2.0, "FibreRequirement"
prob += lpSum([saltPercent[i] * ingredient_vars[i] for i in Ingredients]) <= 0.4, "SaltRequirement"
prob += lpSum([Bvitug[i] * ingredient_vars[i] for i in Ingredients]) >= 0.08, "BvitRequirement"
prob.solve()

print("Status:", LpStatus[prob.status])

procenty = ""

for i in prob.variables():
    if i.varValue > 0:
      procenty += i.name + " - " + str(i.varValue)+ "%, " 

procenty = procenty[:-2]

print("Końcowy produkt będzie kosztować",value(prob.objective),"$. Bedzie miał w składzie:",procenty)
```
