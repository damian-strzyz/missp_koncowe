# **Zadanie**
Zadanie optymalizacyjne polegające na maksymalizacji zysku produkcji okien drewnianych i aluminiowych

*   Stworzenie zadania optymalizacyjnego polegającego na maksymalizacji
*   Zainicjalizowanie zmiennych jako rodzaje produkowanych okien 
*   Określenie celu maksymalizacji
*   Wprowadzenie ograniczeń wynikających z możliwości pracy fabryk
*   Rozwiązanie problemu
*   Wypisanie rozwiązania i wartości zmiennych




```
from pulp import *

problem_okna = LpProblem("Problem_okna", LpMaximize)

okna_alu = LpVariable("okna_alu",lowBound=0, cat= 'Continuous')
okna_drew = LpVariable("okna_drew",lowBound=0, cat= 'Continuous')

problem_okna += 5000*okna_drew+3000*okna_alu,"ZYSK"
problem_okna += 1*okna_alu <= 4, "f1"
problem_okna += 2*okna_drew <= 12, "f2"
problem_okna += 3*okna_alu + 2*okna_drew <= 18, "f3"

problem_okna.solve()
print(LpStatus[problem_okna.status])
print(value(problem_okna.objective))

for variable in problem_okna.variables():
    print("{} = {}".format(variable.name, variable.varValue))
```
