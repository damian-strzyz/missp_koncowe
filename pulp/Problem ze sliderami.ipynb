Zadanie optymalizacyjne polegające na maksymalizacji zysku produkcji okien drewnianych i aluminiowych, podczas gdy cena i czas pracy fabryk jest regulowana za pomocą sliderów

*   Rozwiązanie problemu
*   Wypisanie rozwiązania i wartości zmiennych



```
from __future__ import print_function
from pulp import *
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

cena_drew_s = IntSlider(min=0,max=10000,value=5000,describtion="cena_drewna",style=style)
cena_alu_s = IntSlider(min=0,max=10000,value=3000,describtion="cena_aluminium",style=style)
czas_f1s = IntSlider(min=0,max=24,value=4,describtion="czas_farbyki1",style=style)
czas_f2s = IntSlider(min=0,max=24,value=12,describtion="czas_farbyki2",style=style)
czas_f3s = IntSlider(min=0,max=24,value=18,describtion="czas_farbyki3",style=style)

def produkcja_okien(cena_drew = 5000, cena_alu = 3000, czas_f1 = 4, czas_f2 = 12, czas_f3 = 18):

  problem_okna = LpProblem("Problem_okna", LpMaximize)

  okna_alu = LpVariable("okna_alu",lowBound=0, cat= 'Continuous')
  okna_drew = LpVariable("okna_drew",lowBound=0, cat= 'Continuous')

  problem_okna += cena_drew*okna_drew+cena_alu*okna_alu,"ZYSK"
  problem_okna += 1*okna_alu <= czas_f1, "f1"
  problem_okna += 2*okna_drew <= czas_f2, "f2"
  problem_okna += 3*okna_alu + 2*okna_drew <= czas_f3, "f3"

  problem_okna.solve()
  print(value(problem_okna.objective))
  for variable in problem_okna.variables():
    print("{} = {}".format(variable.name, variable.varValue))

  shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in problem_okna.constraints.items()]
  print("*"*10+"Shadow Prices"+"*"*10)
  print(pd.DataFrame(shadows))

interact(produkcja_okien,cena_drew=cena_drew_s,cena_alu=cena_alu_s, czas_f1 = czas_f1s, czas_f2 = czas_f2s, czas_f3 = czas_f3s)

```
